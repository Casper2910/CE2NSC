For loop no numpy:

```python
import numpy as np
import random
from math import sqrt, pi

def calc_distance_to_zero(x, y):
    distance = sqrt(x**2 + y**2)

    return distance

def monte_carlo(runs = 10_000_000):
    inside = 0
    for i in range(runs):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)

        distance = calc_distance_to_zero(x, y)

        if distance <= 1:
            inside += 1

    estimate = (4 * inside) / (runs)
    
    return estimate

%timeit monte_carlo()
monte_carlo()
```

vectorized:

```python
import numpy as np

def monte_carlo_vec(runs=10_000_000):
    x = np.random.random(runs)
    y = np.random.random(runs)

    inside = np.sum(x*x + y*y <= 1.0)

    return 4 * inside / runs

%timeit monte_carlo_vec()
monte_carlo_vec()
```

with numba:

```python
import numpy as np
from numba import njit

@njit
def monte_carlo_numb(runs=10_000_000):
    inside = 0
    for i in range(runs):
        x = np.random.random()
        y = np.random.random()

        distance = sqrt(x**2 + y**2)

        if distance <= 1:
            inside += 1

    estimate = (4 * inside) / (runs)
    
    return estimate

%timeit monte_carlo_numb()
monte_carlo_numb()
```
