with numba:

```python
from numba import njit

import time
import numpy as np

@njit
def non_unrolled(x, y):
    i = 0
    max = len(x)
    result = 0.0
    while i < max:
        result += (x[i]*y[i])
        i += 1
    return result

@njit
def unrolled(x, y):
    i = 0
    max = len(x)
    sum1 = 0.0
    sum2 = 0.0
    while i < max:
        sum1 += (x[i]*y[i])
        sum2 += (x[i+1]*y[i+1])
        i += 2
    return sum1 + sum2

@njit
def more_unrolled(x, y):
    i = 0
    max = len(x)
    sum1 = 0.0
    sum2 = 0.0
    sum3 = 0
    sum4 = 0
    while i < max:
        sum1 += (x[i]*y[i])
        sum2 += (x[i+1]*y[i+1])
        sum3 += (x[i+2]*y[i+2])
        sum4 += (x[i+3]*y[i+3])
        i += 4
    return sum1 + sum2 #+ sum3 + sum4

x = np.random.normal(size=int(1e6))
y = np.random.normal(size=int(1e6))

%timeit non_unrolled(x,y)
%timeit unrolled(x,y)
%timeit more_unrolled(x,y)
```
